```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    sns.histplot(y=data[column], kde=True)
    plt.title(f'Histogram of Variabel {column}')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[column])
    plt.title(f'Box Plot of {column}')
    
    plt.show()

```

```python
if 'Variabel 10' in data.columns and 'Variabel y' in data.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data['Variabel 10'], y=data['Variabel y'])
    plt.title('Scatter Plot between Variable 10 and Variable Y')
    plt.show()
else:
    print("Salah satu atau kedua kolom tidak ada dalam DataFrame.")
```

LAPORAN

1. Eksplorasi Awal
Setelah mengimpor dataset dan melakukan eksplorasi awal, kami mendapatkan informasi berikut:

Baris Pertama dan Terakhir: Dataset mencakup data dengan beberapa variabel yang relevan untuk analisis lebih lanjut. Melihat beberapa baris pertama dan terakhir membantu dalam memahami struktur umum data dan format.

Missing Values: Dataset memiliki beberapa nilai yang hilang. Kolom-kolom dengan nilai yang hilang perlu ditangani melalui pengisian nilai atau penghapusan baris/kolom.

Statistik Deskriptif: Data memiliki rentang nilai yang bervariasi di setiap kolom numerik. Statistik deskriptif seperti mean, median, dan standar deviasi memberikan wawasan awal tentang distribusi data.

2. Statistik Deskriptif
Mean dan Median: Terdapat perbedaan antara mean dan median untuk beberapa variabel, menunjukkan kemungkinan distribusi yang tidak simetris.

Mode: Variabel memiliki nilai mode tertentu yang sering muncul, yang bisa menunjukkan kecenderungan data.

Standar Deviasi dan Variansi: Variansi dan standar deviasi menunjukkan sejauh mana nilai-nilai dalam dataset menyebar dari rata-rata. Variansi yang tinggi menunjukkan penyebaran data yang luas.

Skewness dan Kurtosis:

Skewness: Mengindikasikan distribusi yang miring. Skewness positif menunjukkan ekor distribusi yang lebih panjang di sisi kanan, sementara skewness negatif menunjukkan ekor di sisi kiri.
Kurtosis: Mengukur puncak distribusi. Kurtosis positif menunjukkan distribusi dengan puncak tajam, sedangkan kurtosis negatif menunjukkan puncak yang lebih datar.
3. Korelasi Antar Variabel
Analisis korelasi menunjukkan adanya hubungan kuat antara beberapa variabel. Korelasi positif atau negatif yang tinggi dapat menunjukkan adanya keterkaitan yang signifikan, yang penting untuk analisis lebih lanjut dan pembuatan model.

4. Visualisasi Data
Histogram: Histogram menunjukkan distribusi nilai untuk setiap variabel. Beberapa variabel menunjukkan distribusi yang tidak normal, dengan puncak atau ekor yang jelas.

Box Plot: Box plot membantu mengidentifikasi outliers (nilai ekstrim) di dalam data. Outliers ditemukan di beberapa variabel, menunjukkan nilai-nilai yang jauh dari nilai rata-rata.

Scatter Plot: Scatter plot antara variabel dengan korelasi tinggi menunjukkan pola hubungan antara variabel tersebut. Ini membantu dalam memahami keterkaitan dan potensi dependensi antara variabel.

Penemuan Signifikan
Distribusi yang Tidak Normal: Beberapa variabel menunjukkan distribusi yang miring atau tidak simetris, yang mungkin memerlukan transformasi untuk analisis lebih lanjut.

Korelasi Kuat: Variabel tertentu menunjukkan korelasi kuat, yang mungkin menunjukkan hubungan signifikan yang perlu dieksplorasi lebih dalam, terutama dalam konteks model prediktif.

Outliers: Ditemukan beberapa outliers dalam dataset, yang mungkin mempengaruhi hasil analisis statistik dan perlu ditangani sesuai kebutuhan.

Missing Values: Terdapat nilai yang hilang di beberapa kolom. Pengisian nilai atau penghapusan baris dengan nilai hilang harus dipertimbangkan untuk memastikan kualitas data yang baik.
